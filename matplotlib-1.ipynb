### Line plot

* a lineplot graph is a graph that shows frequency of data along a number line
* it is a simple way to organize data
* plt.plot()

```python
x_axis=[10,12,14,16]
y_axis=[5,6,7,8]
plt.plot(x_axis,y_axis,label='Line')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Line Graph')
plt.legend(loc='lower right')
plt.show()
```

```python
help(plt.legend)
```

```python
### Functional approach

import numpy as np
x=np.linspace(0,10,20)
y=x**2
plt.plot(x,y)
```

```python
x=np.linspace(0,5,20)
x
```

```python
plt.plot(x,x**2,color='red',label='Line1')
plt.plot(x,x**3,color='green',label='Line2')
plt.plot(x,x**4,color='yellow',label='Line3')
plt.legend()
```

```python
x1=np.arange(100)
x1
```

```python
plt.plot(x1,x1+10,color='r',linestyle='--')
plt.plot(x1,x1+15,color='m',linestyle=':')
plt.plot(x1,x1+20,color='y',linestyle='-.')
plt.plot(x1,x1+25,color='b',linestyle='-')
```

```python
x2=np.linspace(10,30,50)
x2
```

```python
plt.plot(x2,np.sin(x2),color='r',marker='o',markersize=10,markerfacecolor='y')
plt.plot(x2,np.cos(x2),color='g',marker='D',markersize=10,markerfacecolor='c')
```
